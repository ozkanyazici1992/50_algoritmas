```python
def locker(x):
    locker = [0 for i in range (x)]
    for i in range(x):
        for j in range(i,x,i+1):
            if locker[j] == 0:
                locker[j]=1
            elif locker[j]==1:
                locker[j]=0
    return locker
locker(20)
```

```python
def locker(x):
    locker = [0 for i in range (x)]
    for i in range(x):
        for j in range(i,x,i+1):
            if locker[j] == 0:
                locker[j]=1
            elif locker[j]==1:
                locker[j]=0
    return locker.count(1)
locker(20)
```
